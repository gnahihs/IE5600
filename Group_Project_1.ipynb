*   **Computing the log returns for the following:**

1.   Individual Stock (Apple, JNJ)

Libraries Used: numpy, matplotlib




```python
#NOT for SLIDE

#start end
start = datetime(2019, 1, 1)
end = datetime(2020, 9, 6)
 
 #for the returns graphs
  
df = yf.download(['^GSPC', 'AAPL', 'FB','AMZN','GOOG','NFLX','NVS','ABT','PFE','MRK','JNJ'], 
                 start=start, 
                 end=end,
                 progress=False)

df = df[['Adj Close']]
df.columns = df.columns.droplevel(0)

df['aapl_rtn'] = np.log(df.AAPL / df.AAPL.shift(1))
df['jnj_rtn'] = np.log(df.JNJ / df.JNJ.shift(1))
df['medical_rtn'] = np.log(df.NVS / df.NVS.shift(1))*0.2 + np.log(df.ABT / df.ABT.shift(1))*0.2 + np.log(df.PFE / df.PFE.shift(1))*0.2 + np.log(df.MRK / df.MRK.shift(1))*0.2 + np.log(df.JNJ / df.JNJ.shift(1))*0.2
df['tech_rtn'] = np.log(df.AAPL / df.AAPL.shift(1))*0.2 + np.log(df.FB / df.FB.shift(1))*0.2 + np.log(df.AMZN / df.AMZN.shift(1))*0.2 + np.log(df.GOOG / df.GOOG.shift(1))*0.2 + np.log(df.NFLX / df.NFLX.shift(1))*0.2
df['all_rtn'] = np.log(df.AAPL / df.AAPL.shift(1))*0.1 + np.log(df.FB / df.FB.shift(1))*0.1 + np.log(df.AMZN / df.AMZN.shift(1))*0.1 + np.log(df.GOOG / df.GOOG.shift(1))*0.1 + np.log(df.NFLX / df.NFLX.shift(1))*0.1 + np.log(df.NVS / df.NVS.shift(1))*0.1 + np.log(df.ABT / df.ABT.shift(1))*0.1 + np.log(df.PFE / df.PFE.shift(1))*0.1 + np.log(df.MRK / df.MRK.shift(1))*0.1 + np.log(df.JNJ / df.JNJ.shift(1))*0.1


fig, ax = plt.subplots(1, 1, figsize=(24, 5))

df.aapl_rtn.plot(color = 'darkturquoise')
ax.set(xlabel = 'Date', 
          ylabel = 'Log returns (%)', 
          title = 'AAPL Returns time series')

fig, ax = plt.subplots(1, 1, figsize=(24, 5))
df.jnj_rtn.plot(color = 'red')
ax.set(xlabel = 'Date', 
          ylabel = 'Log returns (%)', 
          title = 'JNJ Returns time series')
```

```python
#FOR SLIDE: JEFFERY
#Graphs on Returns
plt.figure(figsize=(24,8))

df.medical_rtn.plot(color = 'green')
df.tech_rtn.plot(color = 'purple')
df.all_rtn.plot(color = 'black')

plt.xlabel('Date')
plt.ylabel('Log Returns %')
plt.title('Returns time series')
plt.legend(['5 Medical Stocks', 'FAANG Stocks' ,'All 10 Stocks'])
```
